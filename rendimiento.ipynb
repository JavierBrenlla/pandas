```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("2022-09-16/rendimiento_formateado.csv")
```

### Tiempo medio por URL

```python
urls = df.groupby("URL").sum()

print("MEDIA: " + str(urls.mean()["TIME"]) + " ms")
print("DESVIACION TIPICA: " + str(urls.std()["TIME"]) + " ms")
```

### Percentil 25 de las URLS

```python
percentile = np.percentile(urls["TIME"], 25)
print(str(percentile) + " ms")
```

### Percentil 50 de las URLS

```python
percentile = np.percentile(urls["TIME"], 50)
print(str(percentile) + " ms")
```

### Percentil 75 de las URLS

```python
percentile = np.percentile(urls["TIME"], 75)
print(str(percentile) + " ms")
```

### Tiempo medio por seccion

```python
groupByUrls = df.groupby(["URL", "SECCION"]).sum()

groupBySeccion = groupByUrls.groupby("SECCION").mean()

desviacionLibra = df[df["SECCION"] == 'Libra']

desviacionDatos = df[df["SECCION"] == 'Datos']

desviacionPermisos = df[df["SECCION"] == 'Permisos']

desviacionRenderizado = df[df["SECCION"] == 'Renderizado']

print("Desviacion tipica de Datos: " + str(desviacionDatos["TIME"].std()) + " ms")
print("Desviacion tipica de Libra: " + str(desviacionLibra["TIME"].std()) + " ms")
print("Desviacion tipica de Permisos: " + str(desviacionPermisos["TIME"].std()) + " ms")
print("Desviacion tipica de Renderizado: " + str(desviacionRenderizado["TIME"].std()) + " ms")

groupBySeccion.plot(kind="bar", yerr=[desviacionDatos["TIME"].std(), desviacionLibra["TIME"].std(), desviacionPermisos["TIME"].std(), desviacionRenderizado["TIME"].std()])
```

### Tiempo medio por vista

```python
groupByUrls = df.groupby(["URL", "VISTA"]).sum()

groupByVista = groupByUrls.groupby("VISTA").mean()

groupByVista

#groupByVista.plot(kind="bar")




groupByUrls = df.groupby(["URL", "VISTA"]).sum()

groupByVista = groupByUrls.groupby("VISTA").mean()

desviacionArticulo = df[df["VISTA"] == 'articulo']

desviacionCategoria = df[df["VISTA"] == 'categoria']

desviacionLibra = df[df["VISTA"] == 'getDatosReferencia']

desviacionGrupo = df[df["VISTA"] == 'grupo']

desviacionHome = df[df["VISTA"] == 'home']

print("Desviacion tipica de Categoria: " + str(desviacionCategoria["TIME"].std()) + " ms")
print("Desviacion tipica de Articulo: " + str(desviacionArticulo["TIME"].std()) + " ms")
print("Desviacion tipica de Libra: " + str(desviacionLibra["TIME"].std()) + " ms")
print("Desviacion tipica de Grupo: " + str(desviacionGrupo["TIME"].std()) + " ms")
print("Desviacion tipica de Home: " + str(desviacionHome["TIME"].std()) + " ms")

groupByVista.plot(kind="bar", yerr=[desviacionCategoria["TIME"].std(), desviacionArticulo["TIME"].std(), desviacionLibra["TIME"].std(), desviacionGrupo["TIME"].std(), desviacionHome["TIME"].std()])
```

### Tiempo de Renderizado por Secciones

```python
renderizado = df[df["SECCION"] == "Renderizado"]

desviacionCategoria = df[(df["VISTA"] == "categoria") & (df["SECCION"] == "Renderizado")]

desviacionGrupo = df[(df["VISTA"] == "grupo") & (df["SECCION"] == "Renderizado")]

desviacionArticulo = df[(df["VISTA"] == "articulo") & (df["SECCION"] == "Renderizado")]

desviacionHome = df[(df["VISTA"] == "home") & (df["SECCION"] == "Renderizado")]

desviacionCategoria["TIME"].std()

groupByCategorias = renderizado.groupby("VISTA").mean()

groupByCategorias.plot(kind="bar", yerr=[desviacionArticulo["TIME"].std(), desviacionCategoria["TIME"].std(), desviacionGrupo["TIME"].std(), desviacionHome["TIME"].std()])

print("Desviacion tipica de Articulo en seccion Renderizado: " + str(desviacionArticulo["TIME"].std()) + " ms")
print("Desviacion tipica de Categoria en seccion Renderizado: " + str(desviacionCategoria["TIME"].std()) + " ms")
print("Desviacion tipica de Grupo en seccion Renderizado: " + str(desviacionGrupo["TIME"].std()) + " ms")
print("Desviacion tipica de Home en seccion Renderizado: " + str(desviacionHome["TIME"].std()) + " ms")
```
